## **Scrapping Movies Data**

````python
# Specifying the URL from which movies related data will be fetched
url='https://www.justwatch.com/in/movies?release_year_from=2000'

# Sending an HTTP GET request to the URL
page=requests.get(url)
# Parsing the HTML content using BeautifulSoup with the 'html.parser'
soup=BeautifulSoup(page.text,'html.parser')
# Printing the prettified HTML content
print(soup.prettify())
````

## **Fetching Movie URL's**

````python
# Write Your Code here
movie_url=[]
# Find the elements that contain the movie URLs
movie_container=soup.find_all('a',class_='title-list-grid__item--link')
# Extract and print the URLs
for movie in movie_container:
       movie_url.append(movie['href'])
print( movie_url)
````

````python
#importing library for joining url
from urllib.parse import urljoin
base_url='https://www.justwatch.com/in/'

movies=[]
for relative_url in movie_url :
    absolute_url = urljoin(base_url, relative_url)
    movies.append(absolute_url)

# Print the list of absolute movie URLs
print(movies)
````

## **Scrapping Movie Title**

````python
# Write Your Code here
movie_title=[]
# Extract and print the movie names
for movie in movies:
  movie_title.append(movie.split('/')[-1])

# Print the extracted movie names
print(movie_title)
````

## **Scrapping release Year**

````python
# Write Your Code here
release_years=[]
for movie_url in movies:
  # Fetch the HTML content of the movie URL using requests
    response=requests.get(movie_url)
    if response.status_code==200:
        # Parse the HTML content with BeautifulSoup
        soup=BeautifulSoup(response.content, 'html.parser')
        try:
            # Find the release year element
            release_year_element=soup.find('span', class_='text-muted')
            release_year=release_year_element.get_text(strip=True).strip("()")
        except AttributeError:
            release_year='N/A'
        release_years.append({'Movie URL': movie_url, 'Release Year': release_year})
    else:
        print(f"Failed to fetch the HTML content for {movie_url}")

# Print the final result
for movie_data in release_years:
    print(movie_data)
````

````python
# Extract release years from the Release_year list
movie_release_years=[year["Release Year"] for year in release_years]

# Display release years
print(movie_release_years)
````

## **Scrapping Genres**

````python
# Write Your Code here
movie_genre=[]
# Fetch the HTML content of the movie URL using requests
for movie_url in movies:
    response=requests.get(movie_url)
    if response.status_code==200:
        soup=BeautifulSoup(response.content, 'html.parser')
        try:
          # Find the movie genre element:
          genres_heading=soup.find('h3', class_='detail-infos__subheading', string='Genres')
          genres_element=genres_heading.find_next_sibling('div', class_='detail-infos__value')
          genres=genres_element.text.strip()
        except AttributeError:
          genres='N/A'
        #append genres in list:
        movie_genre.append(genres)
    else:
        print(f'Failed to fetch the HTML content for {movie_url}')

# Display the final result
print(movie_genre)
````

## **Scrapping IMBD Rating**

````python
# Write Your Code here
movie_rating=[]
# Fetch the HTML content of the movie URL using requests
for movie_url in movies:
    response=requests.get(movie_url)
    if response.status_code==200:
        soup = BeautifulSoup(response.content, 'html.parser')
        # Find the h3 element with the class "detail_infos_subheading" and text "rating"
        rating_heading=soup.find('h3', class_='detail-infos__subheading', string='Rating')
        if rating_heading:
          # If the heading is found, find the next sibling div with class "detail_infos_value"
          rating_element=rating_heading.find_next_sibling('div', class_='detail-infos__value')
          # Check if runtime_element is not None before accessing its text
          if rating_element:
             movie_rating_value=rating_element.text.strip()
          else:
             movie_rating_value='N/A'
        else:
           movie_rating_value='N/A'
        #append in empty list movie_rating:
        movie_rating.append(movie_rating_value)
    else:
        print(f"Failed to fetch the webpage for {movie_url}")

# display result:
print(movie_rating)
````

````python
#getting imdb rating
imdb_rating=[]
for movie in movie_rating:
  imdb_rating.append(movie.split(' ')[0])
print(imdb_rating)
````

## **Fetching Streaming Service Details**

````python
# Write Your Code here
Provider=[]
for movie_url in movies:
    response=requests.get(movie_url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        try:
            # Find the provider element
            provider_element=soup.find('img', class_='offer__icon')
            provider_name=provider_element.get('alt')
        except:
            provider_name='N/A'
        #append in list provider:
        Provider.append( provider_name)
    else:
        print(f"Failed to fetch the webpage for {movie_url}")

#print end result:
print(Provider)
````

````python
#Let's Visvalize it using word cloud
# Importing Libraries
from wordcloud import WordCloud
import matplotlib.pyplot as plt
````

````python
# Combine all streaming service into a single string
streaming_text=' '.join(df_service_counts['Streaming Service'].astype(str))

# Create a WordCloud object
wordcloud=WordCloud(width=800, height=400, background_color='white').generate(streaming_text)

# Display the WordCloud image using matplotlib
plt.figure(figsize=(13, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
````
